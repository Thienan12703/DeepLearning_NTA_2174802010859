```python
#bai 2
import torch
x = torch.tensor([2.0], requires_grad=True)

def f(x):
    return x**3 + 2*x**2 + 5*x + 1

alpha = 0.1
iterations = 10

print("Bt u Gradient Descent")
print("=" * 50)
print(f"Gi tr ban u: x = {x.item():.4f}")
print(f"y ban u = {f(x).item():.4f}")
print()

for i in range(iterations):
    y = f(x)
    
    y.backward()
    
    gradient = x.grad.item()
    
    print(f"Vng lp {i+1}:")
    print(f"  x = {x.item():.6f}")
    print(f"  y = {y.item():.6f}")
    print(f"  o hm dy/dx = {gradient:.6f}")  
    
    
    with torch.no_grad():  
        x -= alpha * gradient
    
    x.grad.zero_()
    
    print(f"  x_new = {x.item():.6f}")
    print("-" * 40)

print("=" * 50)
print(f"Gi tr cui cng: x = {x.item():.6f}")
print(f"Gi tr cui cng: y = {f(x).item():.6f}")


print("\nKim tra bng tnh ton th cng:")
print(f"Hm s: y = x^3 + 2x^2 + 5x + 1")
print(f"o hm: dy/dx = 3x^2 + 4x + 5")
print(f"Ti x = 2: dy/dx = 3*(2)^2 + 4*2 + 5 = 12 + 8 + 5 = 25")
```

```python
# Bai 3
import torch
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
torch.manual_seed(42)

n_samples = 50
x_data = torch.rand(n_samples) * 9 + 1 

noise = torch.randn(n_samples) * 2  
y_data = 3 * x_data + 5 + noise

print("Thng tin tp d liu:")
print(f"S mu: {n_samples}")
print(f"x: min={x_data.min():.2f}, max={x_data.max():.2f}")
print(f"y: min={y_data.min():.2f}, max={y_data.max():.2f}")
print()

w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

print(f"Tham s khi to:")
print(f"w = {w.item():.4f}")
print(f"b = {b.item():.4f}")
print()

alpha = 0.01
n_epochs = 100

loss_history = []
w_history = []
b_history = []

print("Bt u hun luyn...")
print("=" * 60)

for epoch in range(n_epochs):
    y_pred = w * x_data + b
    loss = torch.mean((y_pred - y_data) ** 2)
    loss.backward()
    loss_history.append(loss.item())
    w_history.append(w.item())
    b_history.append(b.item())
    
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1:3d}/{n_epochs}:")
        print(f"  Loss (MSE) = {loss.item():.4f}")
        print(f"  w = {w.item():.4f}, b = {b.item():.4f}")
        print(f"  Gradient w = {w.grad.item():.4f}, Gradient b = {b.grad.item():.4f}")
        print("-" * 40)
    
    with torch.no_grad():  
        w -= alpha * w.grad
        b -= alpha * b.grad
    
    w.grad.zero_()
    b.grad.zero_()

print("=" * 60)
print("Hun luyn hon thnh!")
print()
print(f"Tham s cui cng:")
print(f"w = {w.item():.4f} (gn vi 3)")
print(f"b = {b.item():.4f} (gn vi 5)")
print(f"Loss cui cng: {loss_history[-1]:.4f}")

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].scatter(x_data.numpy(), y_data.numpy(), alpha=0.6, label='D liu thc')
x_line = torch.linspace(1, 10, 100)
y_line = w.item() * x_line + b.item()
axes[0].plot(x_line.numpy(), y_line.numpy(), 'r-', linewidth=2, label=f'y = {w.item():.2f}x + {b.item():.2f}')
axes[0].set_xlabel('Gi hc')
axes[0].set_ylabel('im s')
axes[0].set_title('D liu v ng hi quy')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(range(1, n_epochs + 1), loss_history, 'b-', linewidth=2)
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('MSE Loss')
axes[1].set_title('Loss qua cc epochs')
axes[1].grid(True, alpha=0.3)
axes[1].set_yscale('log') 

axes[2].plot(range(1, n_epochs + 1), w_history, 'r-', label='w', linewidth=2)
axes[2].plot(range(1, n_epochs + 1), b_history, 'g-', label='b', linewidth=2)
axes[2].axhline(y=3, color='r', linestyle='--', alpha=0.5, label='w thc (3)')
axes[2].axhline(y=5, color='g', linestyle='--', alpha=0.5, label='b thc (5)')
axes[2].set_xlabel('Epoch')
axes[2].set_ylabel('Gi tr tham s')
axes[2].set_title('Tham s w v b qua cc epochs')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nSo snh vi cng thc thc t:")
print(f"Cng thc thc t: y = 3x + 5")
print(f"M hnh hc c: y = {w.item():.4f}x + {b.item():.4f}")
print(f"Sai s w: {abs(w.item() - 3):.4f}")
print(f"Sai s b: {abs(b.item() - 5):.4f}")

print("\nD on th:")
test_hours = torch.tensor([2.0, 5.0, 8.0])
predictions = w * test_hours + b
actual = 3 * test_hours + 5

for i in range(len(test_hours)):
    print(f"Gi hc: {test_hours[i]:.1f}h")
    print(f"  D on: {predictions[i].item():.2f} im")
    print(f"  Thc t (khng noise): {actual[i].item():.2f} im")
    print()
```
